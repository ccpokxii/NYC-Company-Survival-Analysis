```python
def plot_confusion_matrix(cm,
                          target_names,
                          title='Confusion matrix',
                          cmap=None,
                          normalize=True):
    """
    given a sklearn confusion matrix (cm), make a nice plot

    Arguments
    ---------
    cm:           confusion matrix from sklearn.metrics.confusion_matrix

    target_names: given classification classes such as [0, 1, 2]
                  the class names, for example: ['high', 'medium', 'low']

    title:        the text to display at the top of the matrix

    cmap:         the gradient of the values displayed from matplotlib.pyplot.cm
                  see http://matplotlib.org/examples/color/colormaps_reference.html
                  plt.get_cmap('jet') or plt.cm.Blues

    normalize:    If False, plot the raw numbers
                  If True, plot the proportions

    Usage
    -----
    plot_confusion_matrix(cm           = cm,                  # confusion matrix created by
                                                              # sklearn.metrics.confusion_matrix
                          normalize    = True,                # show proportions
                          target_names = y_labels_vals,       # list of names of the classes
                          title        = best_estimator_name) # title of graph

    Citiation
    ---------
    http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html

    """
    import matplotlib.pyplot as plt
    import numpy as np
    import itertools

    accuracy = np.trace(cm) / float(np.sum(cm))
    misclass = 1 - accuracy

    if cmap is None:
        cmap = plt.get_cmap('Blues')

    plt.figure(figsize=(15, 15))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()

    if target_names is not None:
        tick_marks = np.arange(len(target_names))
        plt.xticks(tick_marks, target_names, rotation=45)
        plt.yticks(tick_marks, target_names)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]


    thresh = cm.max() / 1.5 if normalize else cm.max() / 2
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        if normalize:
            plt.text(j, i, "{:0.4f}".format(cm[i, j]),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")
        else:
            plt.text(j, i, "{:,}".format(cm[i, j]),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")


    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label\naccuracy={:0.4f}; misclass={:0.4f}'.format(accuracy, misclass))
    plt.show()


from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.datasets import load_digits
from sklearn.model_selection import learning_curve
from sklearn.model_selection import ShuffleSplit


def plot_learning_curve(estimator, title, X, y, axes=None, ylim=None, cv=None,
                        n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    """
    Generate 3 plots: the test and training learning curve, the training
    samples vs fit times curve, the fit times vs score curve.

    Parameters
    ----------
    estimator : object type that implements the "fit" and "predict" methods
        An object of that type which is cloned for each validation.

    title : string
        Title for the chart.

    X : array-like, shape (n_samples, n_features)
        Training vector, where n_samples is the number of samples and
        n_features is the number of features.

    y : array-like, shape (n_samples) or (n_samples, n_features), optional
        Target relative to X for classification or regression;
        None for unsupervised learning.

    axes : array of 3 axes, optional (default=None)
        Axes to use for plotting the curves.

    ylim : tuple, shape (ymin, ymax), optional
        Defines minimum and maximum yvalues plotted.

    cv : int, cross-validation generator or an iterable, optional
        Determines the cross-validation splitting strategy.
        Possible inputs for cv are:
          - None, to use the default 5-fold cross-validation,
          - integer, to specify the number of folds.
          - :term:`CV splitter`,
          - An iterable yielding (train, test) splits as arrays of indices.

        For integer/None inputs, if ``y`` is binary or multiclass,
        :class:`StratifiedKFold` used. If the estimator is not a classifier
        or if ``y`` is neither binary nor multiclass, :class:`KFold` is used.

        Refer :ref:`User Guide <cross_validation>` for the various
        cross-validators that can be used here.

    n_jobs : int or None, optional (default=None)
        Number of jobs to run in parallel.
        ``None`` means 1 unless in a :obj:`joblib.parallel_backend` context.
        ``-1`` means using all processors. See :term:`Glossary <n_jobs>`
        for more details.

    train_sizes : array-like, shape (n_ticks,), dtype float or int
        Relative or absolute numbers of training examples that will be used to
        generate the learning curve. If the dtype is float, it is regarded as a
        fraction of the maximum size of the training set (that is determined
        by the selected validation method), i.e. it has to be within (0, 1].
        Otherwise it is interpreted as absolute sizes of the training sets.
        Note that for classification the number of samples usually have to
        be big enough to contain at least one sample from each class.
        (default: np.linspace(0.1, 1.0, 5))
    """
    if axes is None:
        _, axes = plt.subplots(1, 3, figsize=(20, 5))

    axes[0].set_title(title)
    if ylim is not None:
        axes[0].set_ylim(*ylim)
    axes[0].set_xlabel("Training examples")
    axes[0].set_ylabel("Score")

    train_sizes, train_scores, test_scores, fit_times, _ = \
        learning_curve(estimator, X, y, cv=cv, n_jobs=n_jobs,
                       train_sizes=train_sizes,
                       return_times=True)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    fit_times_mean = np.mean(fit_times, axis=1)
    fit_times_std = np.std(fit_times, axis=1)

    # Plot learning curve
    axes[0].grid()
    axes[0].fill_between(train_sizes, train_scores_mean - train_scores_std,
                         train_scores_mean + train_scores_std, alpha=0.1,
                         color="r")
    axes[0].fill_between(train_sizes, test_scores_mean - test_scores_std,
                         test_scores_mean + test_scores_std, alpha=0.1,
                         color="g")
    axes[0].plot(train_sizes, train_scores_mean, 'o-', color="r",
                 label="Training score")
    axes[0].plot(train_sizes, test_scores_mean, 'o-', color="g",
                 label="Cross-validation score")
    axes[0].legend(loc="best")

    # Plot n_samples vs fit_times
    axes[1].grid()
    axes[1].plot(train_sizes, fit_times_mean, 'o-')
    axes[1].fill_between(train_sizes, fit_times_mean - fit_times_std,
                         fit_times_mean + fit_times_std, alpha=0.1)
    axes[1].set_xlabel("Training examples")
    axes[1].set_ylabel("fit_times")
    axes[1].set_title("Scalability of the model")

    # Plot fit_time vs score
    axes[2].grid()
    axes[2].plot(fit_times_mean, test_scores_mean, 'o-')
    axes[2].fill_between(fit_times_mean, test_scores_mean - test_scores_std,
                         test_scores_mean + test_scores_std, alpha=0.1)
    axes[2].set_xlabel("fit_times")
    axes[2].set_ylabel("Score")
    axes[2].set_title("Performance of the model")

    return plt
```

## Clean raw data <a id='clean'></a>   

[Return to top](#return)

#### Clean raw business data
* Remove non-NYC businesses
* Keep columns that are relevant to the problem of interest
* Check selection bias
* Drop rows with NaN values

```python
# clean raw business data

# only keep business who is in the NY state
df_business = raw_business_data[raw_business_data['Address State'] == 'NY']

# only keep columns that are relevant to the problem of interest
df_business = df_business[['License Expiration Date','License Status', 'License Creation Date',
                           'Industry','Address ZIP','Longitude', 'Latitude','Borough Code']]

# plot distribution after dropping to check selection bias
df=df_business[df_business.isnull().any(axis=1)]
print('Before drop:',len(df_business))
print('After drop:',len(df))

plt.figure()
df.groupby((pd.to_datetime(df['License Creation Date']).dt.year)).count().plot(kind="bar")
plt.show()

plt.figure()
df_business.groupby((pd.to_datetime(df_business['License Creation Date']).dt.year)).count().plot(kind="bar")
plt.show()

# drop rows with NaN values
df_business = df_business.dropna()
df_business
```

#### Clean raw nyc data
* Only keep the columns that are relevant to the problem of interest.
* Create more useful features from current columns


```python
# clean raw nyc data

# only keep the relevant columns 
df_nyc = raw_nyc_data.loc[:,['Zip Code','Congressional District','Population','Business Annual Payroll',
                       'Female Population','Median Age','Households','Income Per Household', 
                       'Number of Businesses','White Population']]

# add new feature columns, since percentages are standardize and useful for comparison
df_nyc.loc[:,'female_percentage'] = df_nyc.loc[:,'Female Population']/df_nyc.loc[:,'Population']
df_nyc.loc[:,'colored_percentage'] = 1-df_nyc.loc[:,'White Population']/df_nyc.loc[:,'Population']

# drop the useless columns after calculation
df_nyc = df_nyc.drop(['Female Population','White Population'],axis=1)

df_nyc
```

#### Merge two dataframe by column ZIP

```python
# merge two dataframes by the intersection of column ZIP

# rename the zip code column in nyc data to match the name in business data
df_nyc.rename(columns = {'Zip Code':'Address ZIP'}, inplace = True)

# asert integer type for both zip code columns
df_business.loc[:,'Address ZIP'] = df_business['Address ZIP'].astype(int)
df_nyc.loc[:,'Address ZIP'] = df_nyc['Address ZIP'].astype(int)

# do the merge, 'inner' uses the intersection of zip codes that both dataframes have
df_com = pd.merge(df_business, df_nyc, how='inner',on='Address ZIP')

# drop rows with NaN values
df_com=df_com.dropna()

# 'reindex' dataframe so indices are successive and do not skip over dropped values, allow for arithemetic later
# note: do not use the .reindex function!
df_com.index = ( np.arange(df_com.shape[0]) )

df_com.columns
```

## Feature processing <a id='feature'></a>   

[Return to top](#return)

#### One-hot encode on industry type

```python
df_com['Industry'].value_counts()
```

```python
# Transform industry-feature using one-hot encoding

industry_dums = pd.get_dummies(df_com['Industry'])

# we have 47 different industries to begin with, so 47 dummy-columns
display(industry_dums)

# add dummys to the main dataframe
df_com = pd.concat([df_com,industry_dums],axis=1)
```

#### Create target variables
* Survival Analysis: `life_span` and `right_censor`
* Multiclass classification: `lifeclass`

```python
# calculate life_span and right_censor

# convert the dates from string objects to datetime objects for easy manipulation
df_com.loc[:,'License Expiration Date'] = pd.to_datetime(df_com['License Expiration Date'])
df_com.loc[:,'License Creation Date'] = pd.to_datetime(df_com['License Creation Date'])

# set today as a datetime object
today = pd.to_datetime(pd.DataFrame({'year': [2019],'month': [11],'day': [22]}))

# make a series of today values for subtraction with another series later
today_series = pd.to_datetime(
    pd.Series(['11/22/2019'] * df_com.shape[0])
)

# add a creation year column
df_com['Creation Year'] = (df_com['License Creation Date'].apply(lambda x: x.year)).astype(int)

# add the censor column
    # if license expires after today, business is alive, assign 0
    # everywhere else, if license expired before today, business has died and event has occurred, assign 1
df_com['right_censor'] = np.where(df_com['License Expiration Date'].values > today.values,0,1)

# add the lifespan column which is in the unit of days
df_com['lifespan'] = (df_com['License Expiration Date']-df_com['License Creation Date']).dt.days

# if business has not died, recalculate lifespan as today - creation date
df_com['lifespan'] = np.where(df_com['right_censor'] == 0, 
                              (today_series-df_com['License Creation Date']).dt.days, 
                              df_com['lifespan'])
```

```python
# Create second target variable class
df_com['lifeclass'] = df_com['lifespan'].apply(lambda x: np.round(x/1095) ) 

df_com=df_com[df_com['lifeclass'] != -1]
df_com=df_com[df_com['lifeclass'] != -2]
df_com['lifeclass'] = np.where(df_com['lifeclass'] >= 7, 7, df_com['lifeclass'])
df_com['lifeclass'] = np.where(df_com['License Expiration Date'].values > today.values,-1,df_com['lifeclass'])


values = df_com['lifeclass'].value_counts()

plt.figure()
objects=['open','<3','3-6','6-9','9-12','12-15','15-18','18-21','>21']
y_pos = np.arange(len(objects))
plt.bar(y_pos,values)
locs, labels = plt.xticks()
ax = [str(i) for i in objects]
plt.xticks(np.arange(9), ax)
plt.xlabel('Lifespan Intervals')
plt.title('Lifespan in 9 classes distribution')

plt.show()
print(values)
```

#### Generate clean data for modeling
* Drop non numerical columns
* Data normalization

```python
# get rid of the first 9 columns, no longer needed, but added back longitude and latitude
df_model = df_com.iloc[:, 9:]

df_model['Longitude'] = df_com['Longitude']
df_model['Latitude'] = df_com['Latitude']

# drop a dummy industry column to remove high collinearity
# let's drop the industry with the fewest instances
df_model = df_model.drop('Ticket Seller Business', 1)

df_model
```

```python
# Once we split the training and testing set, test set should not be touched.
# !!!!All the modelling should be performed on the training set and test on the validation set!!!!

from sklearn.model_selection import train_test_split
from sklearn import preprocessing

train,test = train_test_split(df_model, train_size=0.75,test_size=0.25, random_state=1)
```

```python
# normalize all numeric features to keep coefficients the same magnitudes
# note: you want to normalize after splitting to avoid data leakage
# becaues normalizing with the entire dataset will learn information from the test set's true target values
# when you should not have any informaion about the test target values

train_norm = train.copy()
test_norm = test.copy()

#Select numerical columns which needs to be normalized
cat_col = industry_dums.columns.tolist()
cat_col.extend(['right_censor','lifeclass'])
print(cat_col)
num_col = list(set(train.columns) - set(cat_col))
train_num = train[num_col]
test_num = test[num_col]

# Normalize Training Data 
std_scale = preprocessing.StandardScaler().fit(train_num)
train_norm_num = std_scale.transform(train_num)

#Converting numpy array to dataframe
train_norm_col = pd.DataFrame(train_norm_num, index=train_num.index, columns=train_num.columns) 
train_norm.update(train_norm_col)
print (train_norm.head())

# Normalize Testing Data by using mean and SD of training set
test_norm_num = std_scale.transform(test_num)
test_norm_col = pd.DataFrame(test_norm_num, index=test_num.index, columns=test_num.columns) 
test_norm.update(test_norm_col)
print (test_norm.head())
```

### K-M for different cohorts based on creation year

```python
# sort all the creation year from earliest to latest
create_year_arr = np.sort(train['Creation Year'].unique())
create_year_arr
```

```python
# create a figure and a axes
my_fig = plt.figure(num = 1, figsize = (16,8))
my_ax = my_fig.gca( )

# instantiate another K-M class to create an object for a K-M model
kmf1 = KaplanMeierFitter() 

# separate cohorts based on create year and K-M fit each cohort

years = []
median_survival_time = []

for year in create_year_arr:
# get each creation year    
    # make boolean mask for filtering
    mask = (train['Creation Year'] == year)
    
    # only get instances in that creationg year category
    train_year = train[mask]
    
    # fit the current cohorts data to train the model
    kmf1.fit(train_year['lifespan'], event_observed = train_year['right_censor'], label = year)
    kmf1.plot(ax = my_ax, ci_show=False)
    # ci_show is meant for Confidence interval, which creates a shaded region around your line
    
    years.append(year)
    median_survival_time.append(kmf1.median_survival_time_)

# display the median survival time for each cohort
med_for_yr = pd.DataFrame(median_survival_time, index = years)
med_for_yr.rename(columns = {0:'median survival time'}, inplace = True)
med_for_yr
```

## Model 3: Multiclass Classification<a id='multi'></a>   

[Return to top](#return)

### Data preparation for modeling

```python
train_norm_mc = train_norm.drop(['lifespan','right_censor'],1)
test_norm_mc = test_norm.drop(['lifespan','right_censor'],1)

from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import RandomOverSampler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

## oversampling
ros = RandomOverSampler(random_state=42)
X_train, Y_train = ros.fit_resample(train_norm_mc.drop('lifeclass',1), train_norm_mc['lifeclass'])

np.unique(Y_train, return_counts=True)
X_train
```

### Model Selection

* Algorithms: Random Forest, Decision Tree, K-nearest Neighbors, Logistic Regression
* Evaluation Metrics: Confusion Matrix, Precision, Recall, F1-score

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression

models = [
    RandomForestClassifier(n_estimators=200, max_depth=3, random_state=0),
    DecisionTreeClassifier(criterion='entropy'),
#    LinearSVC(),
#    MultinomialNB(),
    KNeighborsClassifier(),
    LogisticRegression(random_state=0),
]
CV = 5
cv_df = pd.DataFrame(index=range(CV * len(models)))
entries = []
for model in models:
    model_name = model.__class__.__name__
    accuracies = cross_val_score(model, X_train, Y_train, scoring='accuracy', cv=CV)
    for fold_idx, accuracy in enumerate(accuracies):
        entries.append((model_name, fold_idx, accuracy))
    cv_df = pd.DataFrame(entries, columns=['model_name', 'fold_idx', 'accuracy'])
    
import seaborn as sns
plt.figure(figsize=(8,6))
ax = sns.boxplot(x='model_name', y='accuracy', data=cv_df)
sns.stripplot(x='model_name', y='accuracy', data=cv_df, 
              size=8, jitter=True, edgecolor="gray", linewidth=3)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=30
)
plt.show()
```

```python
cv_df.groupby('model_name').accuracy.mean()
```

```python
dt= DecisionTreeClassifier(max_depth=50, random_state=0)
dt.fit(X_train,Y_train)
Y_predict_dt = dt.predict(test_norm_mc.drop('lifeclass',1))

knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train,Y_train)
Y_predict_knn = knn_classifier.predict(test_norm_mc.drop('lifeclass',1))

rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train,Y_train)
rf_predictions = rf_classifier.predict(test_norm_mc.drop('lifeclass',1))

lg_classifier = LogisticRegression()
lg_classifier.fit(X_train,Y_train)
lg_predictions = lg_classifier.predict(test_norm_mc.drop('lifeclass',1))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import numpy as np


cm = confusion_matrix(test_norm_mc['lifeclass'],Y_predict_dt)
class_names = [-1,0,1,2,3,4,5,6,7,8]
disp = plot_confusion_matrix(cm, target_names=class_names,title= "Confusion Matrix",normalize=False)
plt.show()
print(classification_report(test_norm_mc['lifeclass'],Y_predict_dt))


cm = confusion_matrix(test_norm_mc['lifeclass'],Y_predict_knn)
class_names = [-1,0,1,2,3,4,5,6,7,8]
# disp = plot_confusion_matrix(cm, target_names=class_names,title= "Confusion Matrix",normalize=True)
# plt.show()
print(classification_report(test_norm_mc['lifeclass'],Y_predict_knn))


cm = confusion_matrix(test_norm_mc['lifeclass'],rf_predictions)
class_names = [-1,0,1,2,3,4,5,6,7,8]
# disp = plot_confusion_matrix(cm, target_names=class_names,title= "Confusion Matrix",normalize=True)
# plt.show()
pred_prob = knn_classifier.predict_proba(test_norm_mc.drop('lifeclass',1))
print(classification_report(test_norm_mc['lifeclass'],rf_predictions))

cm = confusion_matrix(test_norm_mc['lifeclass'],lg_predictions)
class_names = [-1,0,1,2,3,4,5,6,7,8]
# disp = plot_confusion_matrix(cm, target_names=class_names,title= "Confusion Matrix",normalize=True)
# plt.show()
pred_prob = lg_classifier.predict_proba(test_norm_mc.drop('lifeclass',1))
print(classification_report(test_norm_mc['lifeclass'],lg_predictions))


```
